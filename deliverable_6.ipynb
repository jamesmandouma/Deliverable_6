```python
import pandas as pd

import numpy as np

import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.drop(labels = 'species', axis = 1, inplace = True)
```

```python
iris
```

```python
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris
```

```python
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris
```

```python
summary_dfs = pd.DataFrame(data = {'mean':[np.mean(iris['sepal_length']), np.mean(iris['sepal_width']), np.mean(iris['sepal_sum']), np.mean(iris['petal_length']), np.mean(iris['petal_width']), np.mean(iris['petal_sum'])],
                                  'std':[np.std(iris['sepal_length']), np.std(iris['sepal_width']), np.std(iris['sepal_sum']), np.std(iris['petal_length']), np.std(iris['petal_width']), np.std(iris['petal_sum'])],
                                  'min':[np.min(iris['sepal_length']), np.min(iris['sepal_width']), np.min(iris['sepal_sum']), np.min(iris['petal_length']), np.min(iris['petal_width']), np.min(iris['petal_sum'])],
                                  'max':[np.max(iris['sepal_length']), np.max(iris['sepal_width']), np.max(iris['sepal_sum']), np.max(iris['petal_length']), np.max(iris['petal_width']), np.max(iris['petal_sum'])]},
                                  index=['Sepal Length', 'Sepal Width', 'Sepal Sum', 'Petal Length', 'Petal Width', 'Petal Sum'])
```
